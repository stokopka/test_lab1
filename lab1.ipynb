```python
import os
import pandas as pd
import re
import requests
import datetime
# GetDataByIndex
def reIndex(index):
    IndexDictionary = {1: 24, 2: 25, 3: 5, 4: 6, 5: 27, 6: 23, 7: 26, 8: 7, 9: 11, 10: 13, 11: 14, 12: 15, 13: 16,
                       14: 17, 15: 18, 16: 19, 17: 21, 18: 22, 19: 8, 20: 9, 21: 10, 22: 1, 23: 3, 24: 2, 25: 4}
    return IndexDictionary[index]


def getDataByIndex(index):
    index = reIndex(index)
    strIndex = str(index)
    filename = "{}_{}.txt".format(index,datetime.date.today())
    if os.path.isfile(filename): 
        print("you have this file")
    else:
        url = "https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={}&year1=1981&year2=2020&type=Mean".format(index)
        vhi_url = str(requests.get(url).content).split('<pre>')[1].split('\\n')[:-1]
        out = open(filename, 'wb')
        for i in range(len(vhi_url)):
            vhi_url[i] = re.sub(r',\s\s|\s\s|\s|,\s', ',', vhi_url[i])
            out.write(str.encode(vhi_url[i][:-1] + '\n'))
            
        out.close()

        print("VHI " + strIndex + " is downloaded...")



    
```
